### Part 1. Importing Libraries and Importing Data

In order to begin analysis, we first need to download the python libraries necessary and load in our data. Our code requires the use of a new library, plotly, which may need to be installed for new users.

```python
#importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#regional data
#water_data = pd.read_csv('JMP_2025_REG')

#installing plotly
#!pip install plotly
#importing plotly
import plotly.express as px
```

```python
#reading and loading all SSA data
angola = pd.read_csv('JMP_2025_AGO_Angola_CSV.csv')
burundi = pd.read_csv('JMP_2025_BDI_Burundi_CSV.csv')
cameroon = pd.read_csv('JMP_2025_CMR_Cameroon_CSV.csv')
cent_african_rep = pd.read_csv('JMP_2025_CAF_Central_African_Republic_CSV.csv')
chad = pd.read_csv('JMP_2025_TCD_Chad_CSV.csv')
dem_rep_congo = pd.read_csv('JMP_2025_COD_Congo_Democratic_Republic_CSV.csv')
gabon = pd.read_csv('JMP_2025_GAB_Gabon_CSV.csv')
equ_guinea = pd.read_csv('JMP_2025_GNQ_Equatorial_Guinea_CSV.csv')
s_sudan = pd.read_csv('JMP_2025_SSD_South_Sudan_CSV.csv')
sao_tome_principe = pd.read_csv('JMP_2025_STP_Sao_Tome_and_Principe_CSV.csv')
```

### Part 2. Isolating/Organizing Data

In this section we are taking each region and seperating the urban and rural data for further analysis.

```python
#isolation based on setting 

angola_rural = angola[angola['Setting'] == 'Rural']
angola_urban = angola[angola['Setting'] == 'Urban']

burundi_rural = burundi[burundi['Setting'] == 'Rural']
burundi_urban = burundi[burundi['Setting'] == 'Urban']

cent_african_rep_rural = cent_african_rep[cent_african_rep['Setting'] == 'Rural']
cent_african_rep_urban = cent_african_rep[cent_african_rep['Setting'] == 'Urban']

cameroon_rural = cameroon[cameroon['Setting'] == 'Rural']
cameroon_urban = cameroon[cameroon['Setting'] == 'Urban']

dem_rep_congo_rural = dem_rep_congo[dem_rep_congo['Setting'] == 'Rural']
dem_rep_congo_urban = dem_rep_congo[dem_rep_congo['Setting'] == 'Urban']

gabon_rural = gabon[gabon['Setting'] == 'Rural']
gabon_urban = gabon[gabon['Setting'] == 'Urban']

equ_guinea_rural = equ_guinea[equ_guinea['Setting'] == 'Rural']
equ_guinea_urban = equ_guinea[equ_guinea['Setting'] == 'Urban']

s_sudan_rural = s_sudan[s_sudan['Setting'] == 'Rural']
s_sudan_urban = s_sudan[s_sudan['Setting'] == 'Urban']

sao_tome_principe_rural = sao_tome_principe[sao_tome_principe['Setting'] == 'Rural']
sao_tome_principe_urban = sao_tome_principe[sao_tome_principe['Setting'] == 'Urban']

chad_rural = chad[chad['Setting'] == 'Rural']
chad_urban = chad[chad['Setting'] == 'Urban']
```

### Part 3: Create Plots

In this section we will make plots of our data now that it is organized.

The first plot is a bar chart of different rural regions during 2025. This is to help visualize the access in the different regions.

The plot utilizes the new library plotly. This allows the user to hover over the bar and see the exact houseld access % of the region. It also has a color gradient scale that indicates the percentage as well.

```python
#extracting 2025 piped water access values for each region
data_2025 = {'Angola': angola_rural.loc[angola_rural['Year'] == 2025, 'Piped'].values[0],
            'Cameroon': cameroon_rural.loc[cameroon_rural['Year'] == 2025, 'Piped'].values[0],
            'Central African Republic': cent_african_rep_rural.loc[cent_african_rep_rural['Year'] == 2025, 'Piped'].values[0],
            'Democratic Republic of the Congo': dem_rep_congo_rural.loc[dem_rep_congo_rural['Year'] == 2025, 'Piped'].values[0],
            'Gabon': gabon_rural.loc[gabon_rural['Year'] == 2025, 'Piped'].values[0],
            'Equatorial Guinea': equ_guinea_rural.loc[equ_guinea_rural['Year'] == 2025, 'Piped'].values[0],
            'South Sudan': s_sudan_rural.loc[s_sudan_rural['Year'] == 2025, 'Piped'].values[0],
            'Sao Tome and Principe': sao_tome_principe_rural.loc[sao_tome_principe_rural['Year'] == 2025, 'Piped'].values[0],
            'Chad': chad_rural.loc[chad_rural['Year'] == 2025, 'Piped'].values[0]}

#converting to DataFrame for plotting
df_2025 = pd.DataFrame(list(data_2025.items()), columns=['Country', 'Piped Water Access (%)'])

#plotting bar chart using plotly
#creating interactive bar chart
fig = px.bar(df_2025,
             x='Country',
             y='Piped Water Access (%)',
             color='Piped Water Access (%)',
             color_continuous_scale='viridis',
             title='Rural Piped Water Access Percentage in 2025',
             text='Piped Water Access (%)')

#customizing plot
fig.update_layout(xaxis_title='Country',
                  yaxis_title='% of households with piped water access',
                  xaxis_tickangle=-45,
                  plot_bgcolor='white',
                  title_x=0.5)

#showing numeric labels on bars
fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')

fig.show()
```

```python
#creating figure
plt.figure(figsize = (12,5))

#data for figure
sns.lineplot(x = angola_rural['Year'], y = angola_rural['Piped'], label = "Angola Rural Piped Water Access Percentage")
sns.lineplot(x = cameroon_rural['Year'], y = cameroon_rural['Piped'], label = "Cameroon Rural Piped Water Access Percentage")
sns.lineplot(x = cent_african_rep_rural['Year'], y = cent_african_rep_rural['Piped'], label = "Central African Republic Rural Piped Water Access Percentage")
sns.lineplot(x = dem_rep_congo_rural['Year'], y = dem_rep_congo_rural['Piped'], label = "Democratic Republic of the Congo Rural Piped Water Access Percentage")
sns.lineplot(x = gabon_rural['Year'], y = gabon_rural['Piped'], label = "Gabon Rural Piped Water Access Percentage")
sns.lineplot(x = equ_guinea_rural['Year'], y = equ_guinea_rural['Piped'], label = "Equitorial Guinea Rural Piped Water Access Percentage")
sns.lineplot(x = s_sudan_rural['Year'], y = s_sudan_rural['Piped'], label = "South Sudan Rural Piped Water Access Percentage")
sns.lineplot(x = sao_tome_principe_rural['Year'], y = sao_tome_principe_rural['Piped'], label = "Sao Tome and Principe Rural Piped Water Access Percentage")
sns.lineplot(x = chad_rural['Year'], y = chad_rural['Piped'], label = "Chad Rural Piped Water Access Percentage")

#customizing plot
plt.ylabel('% of households with piped water access')
plt.grid(True)
plt.title(' Rural Piped Water Access Percentage')
plt.show()
```
